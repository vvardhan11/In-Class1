```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time

%matplotlib inline

from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
import re
import urllib
```

```python
url="https://www.actionnetwork.com/nba/injury-report"
user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
headers={'User-Agent':user_agent,} 

request=urllib.request.Request(url,None,headers) #The assembled request
response = urllib.request.urlopen(request)
soup=bs(response)
Injury_data=[]
title=soup.title
allrows=soup.find_all("tr")
for i in allrows:
    row_list=i.find_all("td")
    dataRow=[]
    for cell in row_list:
        dataRow.append(cell.text)
    Injury_data.append(dataRow)
df=pd.DataFrame(Injury_data)
```

```python
df = df.dropna(how='all')
new_columns = ['Name','Position','Status','Reason','Comment','date']

# assign new column names to the DataFrame
df.columns = new_columns
```

```python
import pandas as pd
import gurobipy as gp
from gurobipy import GRB

# Define the positions we need to fill in the team
positions = ["PG", "SG", "SF", "PF", "C", "G", "F", "UTIL"]

# Define the salary cap
salary_cap = 50000

# Create the model
model = gp.Model("Basketball Team")

# Create variables for each player indicating whether they are on the team or not
team_players = model.addVars(data["Name + ID"], vtype=GRB.BINARY)

# Set the objective function to maximize the total points
model.setObjective(gp.quicksum(data.loc[data["Name + ID"] == player, "AvgPointsPerGame"].iloc[0] * team_players[player] for player in data["Name + ID"]), GRB.MAXIMIZE)

# Add constraint for the salary cap
model.addConstr(gp.quicksum(data.loc[data["Name + ID"] == player, "Salary"].iloc[0] * team_players[player] for player in data["Name + ID"]) <= salary_cap)

# Add constraints for the positions we need to fill
for position in positions:
    if position in "C":
        model.addConstr(gp.quicksum(team_players[player] for player in data.loc[data["Position"] == position, "Name + ID"]) <= 1)
    elif position == "G":
        model.addConstr(gp.quicksum(team_players[player] for player in data.loc[(data["Position"] == "PG") | (data["Position"] == "SG"), "Name + ID"]) <= 3)
    elif position == "F":
        model.addConstr(gp.quicksum(team_players[player] for player in data.loc[(data["Position"] == "SF") | (data["Position"] == "PF"), "Name + ID"]) <= 3)
    elif position == "UTIL":
        model.addConstr(gp.quicksum(team_players[player] for player in data["Name + ID"]) == 8)

# Optimize the model
model.optimize()

# Print the selected players, their positions, salaries, and final points
print("Selected players:")
total_salary = 0
total_points = 0
for player in data["Name + ID"]:
    if team_players[player].x > 0:
        position = data.loc[data["Name + ID"] == player, "Position"].iloc[0]
        salary = data.loc[data["Name + ID"] == player, "Salary"].iloc[0]
        points = data.loc[data["Name + ID"] == player, "AvgPointsPerGame"].iloc[0]
        print(f"{player}: {position} (${salary}): {points} points")
        total_salary += salary
        total_points += points

print(f"\nTotal salary: ${total_salary}")
print(f"Total points: {total_points:.2f}")
```
